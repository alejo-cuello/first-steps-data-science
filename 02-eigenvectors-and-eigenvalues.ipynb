Eigenvectors are not affected by transformations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[3,2],[4,1]])
vector = np.array([[1],[1]])
```

```python
u = X.dot(vector)
print(u)
```

```python
plot_vectors([vector.flatten(), u.flatten()], ["black","grey"])
plt.xlim(-1,6)
plt.ylim(-1,6)
```
