```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use("classic")

image = Image.open("./utils/example-image.jpg")

plt.imshow(image)
```

```python
image_grey_scale = image.convert('LA')
print(image_grey_scale)
```

```python
image_matrix = np.array(list(image_grey_scale.getdata(band=0)))
print(image_matrix)
```

```python
image_matrix.shape = (image_grey_scale.size[1], image_grey_scale.size[0])
print(image_matrix)
```

```python
print(image_matrix.shape)
```

```python
plt.imshow(image_matrix, cmap="gray")
```

```python
image_matrix_2 = image_matrix / 10
plt.imshow(image_matrix_2)
```

The relationship between matrix values is more important than the values themselves

```python
plt.imshow(image_matrix_2,cmap="grey")
```

In machine learning it's useful to have all matrix values between 0 and 1

```python
print(np.max(image_matrix))
print(np.max(image_matrix_2))
```

```python
print(np.min(image_matrix))
print(np.min(image_matrix_2))
```
