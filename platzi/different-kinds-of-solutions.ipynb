```python
import numpy as np
import matplotlib.pyplot as plt
```

Linear system equation with no solution

```python
x = np.arange(-6, 6) # Range to evaluate the function

y_1 = 3*x + 5
y_2 = -x + 3
y_3 = 2*x + 1

# It's an overdetermined system: it has more equations than variables
```

```python
plt.figure()
plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```

Linear system equation with one solution

```python
x = np.arange(-6, 6) # Range to evaluate the function

y_4 = -x + 3
y_5 = 2*x + 1

# It's a determined system: The number of equations is equal to the number of variables

plt.figure()
plt.plot(x,y_4)
plt.plot(x,y_5)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```

Linear system equation with infinite solutions

```python
x = np.arange(-6, 6) # Range to evaluate the function

y_6 = 2*x + 1

# It's an undetermined system: It has less equations than variables

plt.figure()
plt.plot(x,y_6)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```
