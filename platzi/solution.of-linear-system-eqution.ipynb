```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5)
```

```python
y_1 = 3*x+5
y_2 = 2*x+3
```

```python
plt.figure()

plt.plot(x, y_1)
plt.plot(x, y_2)

plt.xlim(-5,5)
plt.ylim(-5,5)

plt.axvline(x=0,color='grey')
plt.axhline(y=0,color='grey')
```

```python
#Matrix system
# -3x + y = 5
# -2x + y = 3

# | -3 1 | | x | = | 5 |
# | -2 1 | | y | = | 3 |

A = np.array([[-3, 1],[-2, 1]])
b = np.array([[5],[3]])

sol_1 = np.array([[-2],[-1]]) # The teacher manually calculates the result

print(A.dot(sol_1))
```

```python
# Applying concepts learned in special-types-of-matrices.ipynb

np.set_printoptions(suppress=True)

# A b = x -> A-1 b = x
print(np.linalg.inv(A).dot(b))
```
