```python
import numpy as np
import matplotlib.pyplot as plt
```

Trigonometric functions

```python
N = 100

def f(x):
    return np.cos(x)

x = np.linspace(-10,10,num=N)
y=f(x)

plt.plot(x,y)
plt.grid()
```

Exponential functions

```python
def f(x):
    return np.exp(x)

y=f(x)

plt.plot(x,y)
plt.grid()
```

Logarithm function

```python
def f(x):
    return np.log2(x)

x=np.linspace(0.001,256,1000)
plt.plot(x,f(x))
plt.grid()
```
