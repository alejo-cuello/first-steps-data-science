```python
import matplotlib.pyplot as plt
import numpy as np
```

Subplot: this library allows you to create multiple graphics

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(1,2,1) #Number of rows, columns, and indexes between rows and columns
plt.plot(x,y)
plt.subplot(1,2,2) #Number of rows, columns, and indexes between rows and columns
plt.hist(x)
plt.show()
```

```python
plt.subplot(1,2,1) #Number of rows, columns, and indexes between rows and columns
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(1,2,2) #Number of rows, columns, and indexes between rows and columns
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1) #Number of rows, columns, and indexes between rows and columns
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(2,1,2) #Number of rows, columns, and indexes between rows and columns
plt.pie(y)
plt.show()
```
