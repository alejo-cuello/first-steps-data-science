```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.histplot(data=tips, x='tip',bins=15,cumulative=False,hue='sex',stat='probability',multiple='stack') #Bins = Contenedores
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',fill=True,bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

One function for all graphics:

```python
sns.displot(data=tips,x='tip',hue='sex',kind='hist')
plt.show()
```
