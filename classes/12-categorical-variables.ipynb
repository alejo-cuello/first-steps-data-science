```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex',dodge=True) #We can also use this chart for numerical variables
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex',dodge=True,color='black')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex',split=True,palette='pastel') #We can also use this chart for numerical variables
plt.show()
```

One function for all graphics:

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex',dodge=True,kind='box',col='time')
plt.show()
```
