```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) #In axes returns an array
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2) #In axes returns an array
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(2, 4) #In axes returns an array
axes[0,0].plot(x,np.cos(x),'b')
axes[0,3].plot(x,np.tan(x),'b')
```

```python
fig, ((ax1,ax2,ax3,ax4), (ax5,ax6,ax7,ax8)) = plt.subplots(2, 4) #In axes returns an array
ax1.plot(x,np.cos(x),'b')
ax6.plot(x,np.tan(x),'b')
fig.tight_layout()
```
