```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2,figsize=(5,5))
axes[0].plot(x,y,label='$sin(x)$')
axes[0].set_title('X-Y Relationship')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y,x)
axes[0].set_title('Y-X Relationship')
axes[0].set_xlabel('Y')
axes[0].set_ylabel('X')
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('Title')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right',bbox_to_anchor=(1,0.2))
```
