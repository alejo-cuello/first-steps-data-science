```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

```python
plt.hist(data,bins=50, histtype='step') #Bins=number of bars
plt.show()
```

```python
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True) #Use property showfliers to hide outliers
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area =  (30 * np.random.rand(N)) ** 2

plt.scatter(x,y, s=area, marker='o',alpha=0.5)
plt.show()
```
