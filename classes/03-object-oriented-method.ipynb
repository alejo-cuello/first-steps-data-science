```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

Object oriented method is more complex than pyplot, but it has a better personalization. An example of personalization could be having multiple scales on one axis

![image](../images/figures-and-axes.png)

```python
fig = plt.figure() #Creating a canvas for the graphics
axes = fig.add_axes([0.1,0.1,0.5,0.9]) #Creating an initial graphic with its initial coordinates, width and height
axes.plot(x,y,'b') 
```

```python
fig = plt.figure() #Creating a canvas for the graphics
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #Creating an different object with its owns properties
axes.plot(x,y,'b') 
axes2.plot(y,x,'r') 
axes2.set_facecolor('gray') 
fig.show()
```
