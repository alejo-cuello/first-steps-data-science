```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available) #Returns all default styles
```

Matlab

```python
plt.style.use('bmh')
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:')
```

Pyplot

```python
plt.style.use('bmh')
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color='red', alpha=0.3, linewidth=5, linestyle=':', marker='o',markersize=8,markerfacecolor='black') #Alpha sets transparency 
ax.plot(x,x+2,color='green')
ax.plot(x,x+3,color='#FF66FF')
ax.plot(x,x+4,color='yellow')
```
