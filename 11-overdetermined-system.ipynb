```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 1000)

y_1 = -4 * x + 3
y_2 = 2 * x + 5
y_3 = -3 * x + 1

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-2,2.5)
plt.ylim(-6,6)

plt.show()
```

```python
matrix = np.array([[4,1],[-2,1],[3,1]]) #We find these values clearing up the unkwnowns
```

```python
matrix_pse = np.linalg.pinv(matrix)
print(matrix_pse)
```

```python
b = np.array([[3], [5], [1]])
```

```python
result = matrix_pse.dot(b)
print(result)
```
