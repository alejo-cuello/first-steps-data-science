```python
A = np.array([[3,7],[5,2]])
```

```python
print("Unit circle:")
plot_matrix(np.eye(2))
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()

print("Transformed Unit circle:")
plot_matrix(A)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
U, D, V = np.linalg.svd(A)
```

```python
print("Unit circle:")
plot_matrix(np.eye(2))
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()

print("First rotation V:")
plot_matrix(V)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()

print("Scale D:")
plot_matrix(np.diag(D).dot(V))
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()

print("Second rotation U:")
plot_matrix(U.dot(np.diag(D).dot(V)))
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Comparing plots:

```python
print("Transformed Unit circle:")
plot_matrix(A)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()

print("Transformed Unit circle (using U,V,D):")
plot_matrix(U.dot(np.diag(D).dot(V)))
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
