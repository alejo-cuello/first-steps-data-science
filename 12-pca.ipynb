When I add more variables to my sample, I need exponentially more variables to keep the statistical significance of the sample

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

x=3*np.random.rand(200)
y=20*x+np.random.randn(200)

x= x.reshape(200,1)
y= y.reshape(200,1)

xy = np.hstack([x,y])
print(xy.shape)
```

```python
plt.plot(xy[:,0],xy[:,1],'.')
```

How much information we will lose if we delete a variable?

```python
xy_centered = xy - np.mean(xy, axis=0)
```

```python
plt.plot(xy_centered[:,0],xy_centered[:,1],'.')
```

![image.png](utils/pca-formula.png)

```python
eigenvalues, eigenvectors = np.linalg.eig(xy_centered.T.dot(xy_centered))
print(eigenvectors)
```

The greatest eigenvector with the greatest eigenvalue give us the direction with max variance
