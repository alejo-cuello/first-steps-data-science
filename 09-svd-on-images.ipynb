```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use("classic")

image = Image.open("./utils/example-image.jpg")

plt.imshow(image)
```

```python
image_grey_scale = image.convert('LA')
image_matrix = np.array(list(image_grey_scale.getdata(band=0)), float)
image_matrix.shape = (image_grey_scale.size[1], image_grey_scale.size[0])

print(image_matrix)
```

```python
U,D,V = np.linalg.svd(image_matrix)
```

```python
print(image_matrix.shape)
print(U.shape)
print(D.shape) #The first singular value contains the most information (with the highest variance)
print(V.shape)
```

```python
image_reconstruction = np.matrix(U[:,:1]) * np.diag(D[:1]) * np.matrix(V[:1,:])
plt.imshow(image_reconstruction, cmap="gray")
```

```python
i = 5
image_reconstruction = np.matrix(U[:,:5]) * np.diag(D[:5]) * np.matrix(V[:5,:])

title = "single values = %s" %i
plt.title(title)
plt.imshow(image_reconstruction, cmap="grey")
```
